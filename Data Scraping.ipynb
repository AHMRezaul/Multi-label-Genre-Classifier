```
X = df['URL']
plot = []
for i in range(len(X)):
  link = X[i]+'plotsummary'
  web_page = bs4.BeautifulSoup(requests.get(link, {}).text, "lxml")
  sub_web_page = web_page.find_all(name="li", attrs={"class": "ipl-zebra-list__item"})
  mx = 0
  target = ""
  for j in range(len(sub_web_page)-1):
    part = sub_web_page[j].p.text
    if(mx < len(part)):
       mx = len(part)
       target = part
  plot.append(target)
plot
```
